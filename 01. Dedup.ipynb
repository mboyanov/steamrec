```python
seen_users = set()
with open("data_dir/train-test-split-v1.json", 'w') as out:
    for line in open("/data/data/steamrec/train-test-split.json"):
        doc = json.loads(line)
        if doc['steam_id'] in seen_users:
            continue
        else:
            seen_users.add(doc['steam_id'])
            out.write(line)
```
