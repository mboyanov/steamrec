```python
item_data = pd.read_json('data_dir/rank_500_it_10_lambda_0.1_alpha_1.0/item_factors/part-00000-ac269d3b-5d22-4b37-944b-afe7ecbea095-c000.json', lines=True).sort_values('id').set_index('id')
len(item_data)
```

```python
import json
with open('data_dir/game2id.json') as f: game2id = json.loads(f.read())

id2game = {y:x for x,y in game2id.items()}
```

```python
def get_most_similar(distances, k = 20):
    ind = np.unravel_index(np.argsort(distances, axis=None), distances.shape)
    return [(ind[0][i] ,ind[1][i]) for i in range(0, 2*k, 2)]
```

# Which are the most similar games? 

```python
user_factors = np.array([np.array(x) for x in data.features])
```

```python
item_factors = np.array([np.array(x) for x in item_data.features])
```

```python
# transform to tsv
# np.savetxt('data_dir/item_factors.tsv',item_factors, delimiter= '\t')
# pd.Series(game2id).reset_index().set_index(0).to_csv('data_dir/game2id.tsv', sep='\t')
```

```python
item_distances = euclidean_distances(item_factors) + np.eye(item_factors.shape[0])
```

```python
most_similar_games = get_most_similar(item_distances)
most_similar_games
```

```python
for pair in most_similar_games:
    display( f'{id2game[pair[0]]} ~ {id2game[pair[1]]}')
```

# Who are the most similar users?

```python
user_distances = euclidean_distances(user_factors) + np.eye(user_factors.shape[0])
```

```python
most_similar_users = get_most_similar(user_distances)
most_similar_users
```

```python
user_data = pd.read_json('data_dir/train-test-split-v2.json', lines=True)
```

```python
for pair in get_most_similar(user_distances, k =5):
    user1games = pd.DataFrame(user_data.iloc[pair[0]]['games']).sort_values('playtime_forever', ascending=False).head(20)[['name', 'playtime_forever']]
    user2games = pd.DataFrame(user_data.iloc[pair[1]]['games']).sort_values('playtime_forever', ascending=False).head(20)[['name', 'playtime_forever']]
    display(pd.concat([user1games, user2games],axis=1))
```

# Which are the most recommended games?

```python
recommendations = user_factors.dot(item_factors.T)
best_recommended = get_most_similar(-recommendations)
```

```python
for p in best_recommended:
    ud = pd.DataFrame(user_data.iloc[p[0]]['train_games']).sort_values('playtime_forever', ascending=False)
    game = id2game[p[1]]
    games = recommendations[p[0]].argsort()[::-1][:20]
    games = [id2game[g] for g in games if g not in ud.game_id.values]
    display(games)
    display(ud.head(20))
```
